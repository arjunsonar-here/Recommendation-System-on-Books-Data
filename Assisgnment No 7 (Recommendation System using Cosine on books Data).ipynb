```python
# Creating a Pivot table
user_df=book.pivot_table(index='userID',columns='title',values='rating').reset_index(drop=True)
userID=user_df.index
title=user_df.columns
user_df
```

```python
# Creating index from UserID.
user_df.index=book.userID.unique()
```

```python
# Filling the NA values with zero
user_df.fillna(0, inplace =True)
user_df
```

```python
# Import libraries to find similarities 
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
# FInding Similarities
user_sim=1-pairwise_distances(user_df,metric='cosine')
user_sim
```

```python
user_sim_df=pd.DataFrame(user_sim)
user_sim_df.index=book.userID.unique()
user_sim_df.columns=book.userID.unique()
user_sim_df
```

```python
# Filling the diagonal cells with zero 
np.fill_diagonal(user_sim,0)
user_sim_df
```

```python
# Most similar users
user_sim_df.idxmax(axis=1)[0:20]
```

```python
count_of_rating=pd.DataFrame(book.groupby('title')['rating'].count().sort_values(ascending=False))
count_of_rating.head()
```

```python
avg_rating=pd.DataFrame(book.groupby('title')['rating'].mean().sort_values(ascending=False))
avg_rating
```

```python
sum_of_rating=pd.DataFrame(book.groupby('title')['rating'].sum().sort_values(ascending=False))
sum_of_rating
```

```python
# Comparing best books
book_rating=sum_of_rating.merge(avg_rating,on='title').merge(count_of_rating,on='title')
book_rating
```

```python
book_rating.columns=['Total Rating','Average Rating','Count of Rating']
book_rating
```
